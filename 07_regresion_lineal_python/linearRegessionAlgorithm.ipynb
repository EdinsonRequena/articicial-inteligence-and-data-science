```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def estimate_b0_b1(x, y):
    
    n = np.size(x)
    
    #  Obtenemos los promedios de X & Y
    m_x, m_y = np.mean(x), np.mean(y)
    
    #  Calcular sumatoria de XY & XX
    sumatory_xy = np.sum((x - m_x) * (y - m_y))
    sumatory_xx = np.sum((x - m_x)**2)
    
    #  Coeficientes de Regresion
    b_1 = sumatory_xy / sumatory_xx
    b_0 = m_y - b_1 * m_x
    
    print(b_1)
    print(b_0)
    return(b_0, b_1)
```

```python
def plot_regression(x, y, b):
    
    plt.scatter(x, y, color = 'g', marker = 'o', s = 30)
    
    y_pred = b[0] + b[1] * x
    plt.plot(x, y_pred, color = 'b')
    
    plt.xlabel('x-independiente')
    plt.ylabel('y-dependiente')
    
    plt.show()
```

```python
def main():
    
    x = np.random.randint(1000, size=1000)
    y = np.random.randint(1000, size=1000)
    
    b = estimate_b0_b1(x, y)

    print(f'los valores de b0 = {b[0]}, b1 = {b[1]}')
    
    plot_regression(x, y, b)
```

```python
if __name__ == '__main__':
    
    main()
```
